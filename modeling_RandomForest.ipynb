```python
lag_1_OPS = []
for i in range(len(regular)):
    data = regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"].iloc[i]-1)]["OPS"]
    if not len(data):
        lag_1_OPS.append(np.nan)
    else:
        lag_1_OPS.append(data.iloc[0])
```

```python
regular["lag_1_OPS"] = lag_1_OPS # 새로운 컬럼 생성
```

```python
lag_1_avg = []
for i in range(len(regular)):
    data = regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"].iloc[i]-1)]["avg"]
    if not len(data):
        lag_1_avg.append(np.nan)
    else:
        lag_1_avg.append(data.iloc[0])
```

```python
regular["lag_1_avg"]=lag_1_avg # 새로운 컬럼 생성(해당 선수의 전년도 칼럼)
```

```python
# for문을 이용해서 변수 만들기 -> lag1
for j in ['avg', 'G', 'AB', 'R', 'H','2B', '3B', 'HR', 'TB', 'RBI', 'SB', 'CS', 'BB', 'HBP', 'SO', 'GDP','SLG', 'OBP', 'E']:
    lag_1_avg = []
    for i in range(len(regular)):
        data = regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"].iloc[i]-1)][j]
        if not len(data):
            lag_1_avg.append(np.nan)
        else:
            lag_1_avg.append(data.iloc[0])
    
    regular["lag_1_"+j]=lag_1_avg
    print(j)
```

```python
# for문을 이용해서 변수 만들기 -> lag2
for j in ['avg', 'G', 'AB', 'R', 'H','2B', '3B', 'HR', 'TB', 'RBI', 'SB', 'CS', 'BB', 'HBP', 'SO', 'GDP','SLG', 'OBP', 'E']:
    lag_1_avg = []
    for i in range(len(regular)):
        data = regular.loc[(regular["batter_name"]==regular["batter_name"].iloc[i])&(regular["year"]==regular["year"].iloc[i]-2)][j]
        if not len(data):
            lag_1_avg.append(np.nan)
        else:
            lag_1_avg.append(data.iloc[0])
    
    regular["lag_2_"+j]=lag_1_avg
    print(j)
```

#### 데이터셋 분할

- train data set : 2017년도 이전까지의 data
- test data set : 2018년도의 data

```python
train = regular.loc[regular["year"]<=2017,]
test = regular.loc[regular["year"]==2018,]
y_train = train["OPS"]
X_train = train[['lag_1_avg', 'lag_1_G', 'lag_1_AB', 'lag_1_R',
       'lag_1_H', 'lag_1_2B', 'lag_1_3B', 'lag_1_HR', 'lag_1_TB', 'lag_1_RBI',
       'lag_1_SB', 'lag_1_CS', 'lag_1_BB', 'lag_1_HBP', 'lag_1_SO',
       'lag_1_GDP', 'lag_1_SLG', 'lag_1_OBP', 'lag_1_E', 'lag_1_OPS']]

y_test = test["OPS"]
X_test = test[['lag_1_avg', 'lag_1_G', 'lag_1_AB', 'lag_1_R',
       'lag_1_H', 'lag_1_2B', 'lag_1_3B', 'lag_1_HR', 'lag_1_TB', 'lag_1_RBI',
       'lag_1_SB', 'lag_1_CS', 'lag_1_BB', 'lag_1_HBP', 'lag_1_SO',
       'lag_1_GDP', 'lag_1_SLG', 'lag_1_OBP', 'lag_1_E', 'lag_1_OPS']]
```

```python
rf=RandomForestRegressor(n_estimators=500)
rf.fit(X_train.fillna(-1), y_train.fillna(-1))
predicted = rf.predict(X_test.fillna(-1))
```

```python
# 평가지표는 RMSE -> MSE의 root
ab = test["AB"]
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test.fillna(-1), predicted,sample_weight=ab)**(1/2)
```

- 타석에 선 횟수가 작으면 해당 row는 유의미성이 떨어짐
- 타석에 선 횟수가 많으면 해당 row의 유의미성이 높아짐  
=> 각 row에 가중치를 둠

```python
rf=RandomForestRegressor(n_estimators=500)
rf.fit(X_train.fillna(-1), y_train.fillna(-1), sample_weight = train["AB"])
predicted = rf.predict(X_test.fillna(-1))
```

```python
# 평가지표는 RMSE -> MSE의 root
ab = test["AB"]
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test.fillna(-1), predicted,sample_weight=ab)**(1/2)
```

- 변동성이 큰 지표(ex. avg)를 예측할 땐 역대 지표를 반영하는 것이 좋음
